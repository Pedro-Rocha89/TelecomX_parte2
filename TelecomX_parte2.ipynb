### Extra√ß√£o do Arquivo Tratado

```python
import pandas as pd
```

```python
# Lendo os dados tratados do desafio TelecomX parte 1
df = pd.read_csv('/content/dados_tratados.csv')
```

### Remo√ß√£o de Colunas Irrelevantes

```python
# Removendo a coluna 'id_cliente', essa coluna n√£o tem relev√¢ncia na predi√ß√£o
df = df.drop(['id_cliente'], axis= 1)
df
```

### Encoding

```python
# Analisando todos os dados por colunas
for i in df.columns:
  print(f'Dados da coluna {i}')
  print(df[i].value_counts())
  print("\n")
```

Observando os dados contidos em cada coluna, chegamos a conclus√£o de que devemos codificar os dados das colunas categ√≥ricas "metodo_pagamento", "tipo_contrato", "servico_internet" e "genero".

Precisamos transformar os dados de string para num√©rico, para que o modelo preditivo possa ser aplicado com sucesso.

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
```

```python
# Realizando a codifica√ß√£o
colunas_categoricas = ['metodo_pagamento', 'tipo_contrato', 'servico_internet', 'genero']
one_hot_enc = make_column_transformer((
    OneHotEncoder(drop= 'if_binary', handle_unknown = 'ignore'),
    colunas_categoricas),
    remainder= 'passthrough', sparse_threshold= 0
)
```

```python
# Aplicando ao DataFrame
df_enc = one_hot_enc.fit_transform(df)
df_enc = pd.DataFrame(df_enc, columns= one_hot_enc.get_feature_names_out())
df_enc
```

```python
# Limpando e padronizando os nomes das colunas
prefixos_para_remover = ['remainder__', 'onehotencoder__']
novos_nomes_colunas = df_enc.columns

for prefixo in prefixos_para_remover:
    novos_nomes_colunas = [col.replace(prefixo, '') for col in novos_nomes_colunas]

df_enc.columns = novos_nomes_colunas

df_enc.columns
```

```python
# Confirmando que todos as colunas foram codificadas
for i in df_enc.columns:
  print(f'Dados da coluna {i}')
  print(df_enc[i].value_counts())
  print("\n")
```

```python
df_enc.to_csv('dados_codificados.csv', index = False)
```

# ü§ñ Modelagem Preditiva

### Separa√ß√£o de Dados

```python
from sklearn.model_selection import train_test_split
```

```python
# Definindo as vari√°veis alvo e explicativas
X = df_enc[[
    'tipo_contrato_Month-to-month',
    'tipo_contrato_Two year',
    'servico_internet_Fiber optic',
    'servico_internet_No',
    'metodo_pagamento_Electronic check',
    'meses_de_servico',
    'gastos_totais'
]]

y = df_enc['cancelou_servico']

oversample = SMOTE()
X_balanceado, y_balanceado = oversample.fit_resample(X,y)

X_treino, X_teste, y_treino, y_teste = train_test_split(
    X_balanceado, y_balanceado,
    test_size=0.3,
    stratify=y_balanceado,
    random_state=42
)
```

### Cria√ß√£o de Modelos

#### √Årvore de Decis√£o

O primeiro modelo selecionado para machine learning ser√° a √Årvore de Decis√£o

O modelo √© de f√°cil interpreta√ß√£o, bom para regras simples e n√£o exige normaliza√ß√£o.

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
arvore = DecisionTreeClassifier(max_depth= 10)
arvore.fit(X_treino, y_treino)
```

```python
y_pred_arvore = arvore.predict(X_teste)
```

```python
y_pred_arvore
```

#### Random Forest

O segundo modelo selecionado para machine learning ser√° o Random Forest

As vantagens desse modelo s√£o que ele √© mais robusto e preciso, reduz overfitting, mostra import√¢ncia das vari√°veis.

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
modelo_rf = RandomForestClassifier(max_depth=10)
modelo_rf.fit(X_treino, y_treino)
```

```python
y_pred_rf = modelo_rf.predict(X_teste)
```

```python
y_pred_rf
```

### Avalia√ß√£o dos Modelos

Criando uma fun√ß√£o que calcule m√©tricas para avalia√ß√£o dos modelos.

```python
# As m√©tricas da fun√ß√£o ser√£o: Acur√°cia, Precis√£o, Recall e F1-score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

def avaliar_modelo(nome, y_real, y_pred):
    print(f"üìä Avalia√ß√£o do Modelo: {nome}")
    print(f"Acur√°cia:  {accuracy_score(y_real, y_pred):.4f}")
    print(f"Precis√£o:  {precision_score(y_real, y_pred):.4f}")
    print(f"Recall:    {recall_score(y_real, y_pred):.4f}")
    print(f"F1-Score:  {f1_score(y_real, y_pred):.4f}")
    print()

    mc = confusion_matrix(y_real, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=mc, display_labels=["N√£o Cancelou", "Cancelou"])
    disp.plot(cmap=plt.cm.Blues)
    plt.title(f'Matriz de Confus√£o - {nome}')
    plt.show()
```

#### Avaliando o modelo √Årvore de Decis√£o

```python
avaliar_modelo('√Årvore de Decis√£o', y_teste, y_pred_arvore )
```

#### Avaliando o modelo Random Forest

```python
avaliar_modelo('Random Forest', y_teste, y_pred_rf)
```

O modelo Random Forest apresentou um desempenho superior em todas as m√©tricas avaliadas.

Acur√°cia: O Random Forest acertou 83.9% das previs√µes no conjunto de teste, enquanto a √Årvore de Decis√£o acertou 82.83%.

Precis√£o: Das previs√µes "Sim" (cancelou) feitas pelo Random Forest, 81.69% estavam corretas, superando os 80.38% da √Årvore de Decis√£o. Isso significa que o Random Forest tem uma taxa ligeiramente menor de falsos positivos.

Recall (Sensibilidade): O Random Forest conseguiu identificar 87.37% de todos os clientes que realmente cancelaram, um pouco melhor que os 86.86% da √Årvore de Decis√£o.

F1-Score: Sendo uma m√©dia harm√¥nica entre precis√£o e recall, o F1-Score do Random Forest (84.43%) tamb√©m √© superior, indicando um melhor equil√≠brio geral entre as duas m√©tricas.

Embora a diferen√ßa n√£o seja gigantesca, o Random Forest √© consistentemente melhor em todos os aspectos, tornando-o o modelo vencedor nesta compara√ß√£o.



#### Analisando overfitting ou underfitting

```python
# Testando o Modelo √Årvore de Decis√£o

# Avalia√ß√£o no conjunto de treino
y_pred_treino_arvore = arvore.predict(X_treino)
acc_treino_arvore = accuracy_score(y_treino, y_pred_treino_arvore)
print(f'Acur√°cia no treino: {acc_treino_arvore:.4f}')

# Avalia√ß√£o no conjunto de teste
y_pred_teste_arvore = arvore.predict(X_teste)
acc_teste_arvore = accuracy_score(y_teste, y_pred_teste_arvore)
print(f'Acur√°cia no teste:  {acc_teste_arvore:.4f}')

if (acc_treino_arvore - acc_teste_arvore) > 0.1:
    print("Ind√≠cio de Overfitting: A diferen√ßa entre a acur√°cia de treino e teste √© significativa.\n")
else:
    print("Modelo parece bem generalizado.\n")
```

```python
# Testando o Modelo Random Forest

# Avalia√ß√£o no conjunto de treino
y_pred_treino_rf = modelo_rf.predict(X_treino)
acc_treino_rf = accuracy_score(y_treino, y_pred_treino_rf)
print(f'Acur√°cia no treino: {acc_treino_rf:.4f}')

# Avalia√ß√£o no conjunto de teste
y_pred_teste_rf = modelo_rf.predict(X_teste)
acc_teste_rf = accuracy_score(y_teste, y_pred_teste_rf)
print(f'Acur√°cia no teste:  {acc_teste_rf:.4f}')

if (acc_treino_rf - acc_teste_rf) > 0.1:
    print("Ind√≠cio de Overfitting: A diferen√ßa entre a acur√°cia de treino e teste √© significativa.\n")
else:
    print("Modelo parece bem generalizado.\n")
```

O max_depth arbitrado nos primeiros testes se mostraram satisfat√≥rios.

**An√°lise de Relev√¢ncia das Vari√°veis**


Ambos os modelos, √Årvore de Decis√£o e Random Forest, utilizam um m√©todo baseado na redu√ß√£o da impureza de Gini para determinar a import√¢ncia de cada vari√°vel. A impureza mede a probabilidade de classificar incorretamente um cliente se ele fosse rotulado aleatoriamente de acordo com a distribui√ß√£o das classes no n√≥ da √°rvore. Uma vari√°vel √© considerada importante se as divis√µes baseadas nela resultam em n√≥s "mais puros" (ou seja, n√≥s onde a maioria dos clientes pertence a uma √∫nica classe, "Cancelou" ou "N√£o Cancelou").

**An√°lise do Modelo Random Forest**


O Random Forest eleva esse conceito. Ele constr√≥i centenas de √°rvores de decis√£o e, para cada vari√°vel, calcula a m√©dia da redu√ß√£o de impureza (Mean Decrease in Impurity) que ela proporciona em todas as √°rvores do modelo. Isso torna a sua avalia√ß√£o de import√¢ncia mais robusta e confi√°vel do que a de uma √∫nica √°rvore.

Analisando o resultado do seu modelo Random Forest, as vari√°veis mais importantes s√£o:

Com uma import√¢ncia de 25.5%, este √© o fator mais decisivo para prever o cancelamento. Clientes com contratos mensais n√£o t√™m a mesma fidelidade de longo prazo e podem cancelar o servi√ßo com muito mais facilidade a qualquer sinal de insatisfa√ß√£o ou ao encontrar uma oferta melhor.

* M√©todo de Pagamento (metodo_pagamento_Electronic check):

Este √© o segundo fator mais importante, com 18.4%. Clientes que utilizam cheque eletr√¥nico como forma de pagamento t√™m uma propens√£o muito maior a cancelar. Isso pode indicar um perfil de cliente menos engajado ou mais propenso a reavaliar os gastos a cada pagamento manual, em contraste com pagamentos autom√°ticos no cart√£o de cr√©dito.

* Gastos Totais (gastos_totais):

Com 16.4% de import√¢ncia, este fator continua sendo crucial. Clientes que gastaram menos no total com a empresa (geralmente por serem clientes mais recentes) t√™m maior probabilidade de sair. A lealdade e o investimento do cliente na empresa ainda n√£o foram consolidados.

* Meses de Servi√ßo (meses_de_servico):

Com 14.9%, este fator est√° diretamente ligado aos gastos totais. Confirma que a dura√ß√£o do relacionamento com o cliente √© um pilar da reten√ß√£o. Quanto menos tempo um cliente est√° na base, maior o risco de cancelamento.

## Relat√≥rio de An√°lise Preditiva de Cancelamento de Clientes


### 1. Introdu√ß√£o

Este relat√≥rio apresenta os resultados de uma an√°lise preditiva sobre os fatores que influenciam o cancelamento de clientes na empresa TelecomX. Utilizando um conjunto de dados fornecido, realizamos o pr√©-processamento e o balanceamento dos dados para treinar e avaliar dois modelos de Machine Learning: √Årvore de Decis√£o e Random Forest. O objetivo principal √© identificar os fatores mais relevantes que levam ao cancelamento e, com base neles, propor estrat√©gias de reten√ß√£o eficazes.



---


### 2. Metodologia

- Pr√©-processamento:

  ‚úî A coluna customerID foi removida por n√£o possuir valor preditivo.

  ‚úî As vari√°veis categ√≥ricas (como Contract, InternetService, etc.) foram transformadas em vari√°veis num√©ricas atrav√©s da t√©cnica de One-Hot Encoding.

- Balanceamento de Dados:

  O conjunto de dados original apresentava um desequil√≠brio, com 73.5% dos clientes n√£o cancelando o servi√ßo e apenas 26.5% cancelando.

  Para evitar que o modelo aprendesse a prever apenas a classe majorit√°ria, aplicamos a t√©cnica SMOTE (Synthetic Minority Over-sampling Technique), que gerou amostras sint√©ticas da classe minorit√°ria (clientes que cancelaram), resultando em um conjunto de dados perfeitamente balanceado (50% para cada classe).

- Sele√ß√£o de Vari√°veis e Modelagem:

  Com base na an√°lise de correla√ß√£o e no conhecimento do neg√≥cio, selecionamos as seguintes vari√°veis como preditoras:

  * meses_de_servico

  * gastos_totais

  * gastos_mensais

  * tipo_contrato_Month-to-month

  * servico_internet_Fiber optic

  * metodo_pagamento_Electronic check

  * tipo_contrato_Two year
